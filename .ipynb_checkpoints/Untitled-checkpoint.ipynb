```python
def lire_fichier(nom_fichier):
    with open(nom_fichier,"r") as f:
        g = gt.Graph()
        vprop = g.new_vertex_property("string") 
        
        contenu=list(f)
        
        nb_sommets=int(re.findall("\d+", contenu[0])[0])
        nb_arcs=int(re.findall("\d+", contenu[1])[0])
        
        g.add_vertex(nb_sommets)
        
        for i in range(2,len(contenu)):
            tmp_l=contenu[i].strip()
            
            if tmp_l[:6]=="sommet":
                g.add_vertex(i)
        gt.graph_draw(g, vertex_text=g.vertex_index)
```

```python
lire_fichier("tests/graph_1_test.txt")
```
