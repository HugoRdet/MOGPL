```python
import numpy as np
import networkx as nx
import re
import matplotlib as plt
```

```python
def afficher_graphe(g):
    options = {
    
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 1,
    "width": 2,
    }
    
    pos = nx.random_layout(g)
    
    edge_weight = nx.get_edge_attributes(g,"weight")
    nx.draw_networkx(g, **options,connectionstyle='arc3, rad = 0.1')
    #nx.draw_networkx_edge_labels(g,, edge_labels = edge_weight)
```

```python
def lire_fichier(nom_fichier):
    with open(nom_fichier,"r") as f:
        contenu=list(f)
        
        nb_sommets=int(re.findall("\d+", contenu[0])[0])
        nb_arcs=int(re.findall("\d+", contenu[1])[0])
        
        g = nx.MultiDiGraph()
        
        
        for i in range(2,len(contenu)):
            tmp_l=contenu[i].strip()
            if tmp_l[:6]=="sommet":
                l_infos= tmp_l[6:].split(' ')
                g.add_node(l_infos[1])
                continue
                
            if tmp_l[:3]=="arc":
                
                l_infos= tmp_l[4:].split(' ')
                u=l_infos[0]
                v=l_infos[1]
                t_tmp=int(l_infos[2])
                lambd=int(l_infos[3])
                g.add_edge(u,v,t=t_tmp,l=lambd)
        l=g.edges(data=True,keys=True) 
        print(l)
        afficher_graphe(g)  
            
        
```
